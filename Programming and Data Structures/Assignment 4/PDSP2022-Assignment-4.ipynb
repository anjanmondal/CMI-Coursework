```python
import pandas as pd
```

```python
students=pd.read_csv('students.csv', sep='~')
instructors=pd.read_csv('instructors.csv', sep='~')
courses=pd.read_csv('courses.csv', sep='~')
grades=pd.read_csv('grades.csv',sep='~')
gradepoint=pd.read_csv('gradepoint.csv',sep='~')
houses=pd.read_csv('houses.csv',sep='~')
```

# Question 1
Course name(s) with highest enrolment: (course name, number of students enrolled)

```python
grad_cour=grades.merge(courses).groupby(by='Name').count()
grad_cour['Number of Students']=grad_cour.Rollno
grad_cour=grad_cour[grad_cour['Number of Students']==grad_cour['Number of Students'].max()]
grad_cour.reset_index(inplace=True)
grad_cour[['Name','Number of Students']]
```

## Question 2
Student(s) enrolled for maximum number of courses: (student name, number of courses enrolled)

```python
grad_stu=grades.merge(students).groupby('Name').count()
grad_stu['Number of Courses Enrolled']=grad_stu.Coursecode
grad_stu=grad_stu[grad_stu['Number of Courses Enrolled']==grad_stu['Number of Courses Enrolled'].max()]
grad_stu.reset_index(inplace=True)
grad_stu[['Name', 'Number of Courses Enrolled']]
```

## Question 3
Student(s) with maximum number of distinct grades across courses: (student name, Distinct Grades)

```python
course_count=grades.merge(students).groupby('Name').count().Coursecode
uniq_grades=grades.merge(students).groupby('Name').nunique().Grade
grad_stu=grades.merge(students, how='inner', on='Rollno').groupby('Name').count()
grad_stu['Number of Courses Enrolled']=course_count
grad_stu['Distinct Grades']=uniq_grades

grad_stu=grad_stu[grad_stu['Distinct Grades']==grad_stu['Distinct Grades'].max()]
grad_stu.reset_index(inplace=True)
grad_stu[['Name', 'Distinct Grades']]
```

## Question 4
GPA of each course: (course name, GPA)


```python
grad_mean=grades.merge(gradepoint, left_on='Grade', right_on='Lettergrade').groupby('Coursecode').mean()
grad_mean.reset_index(inplace=True)

grad_mean=grad_mean.merge(courses)
grad_mean.rename(columns={"Name":"Course Name","Gradepoint":"GPA"},inplace=True)
grad_mean[['Course Name','GPA']]
```

## Question 5
GPA of each student: (student name, GPA)


```python
grad_sum=grades.merge(gradepoint, left_on='Grade', right_on='Lettergrade').groupby('Rollno').sum()
grad_sum.reset_index(inplace=True)
grad_count=grades.merge(gradepoint, left_on='Grade', right_on='Lettergrade').groupby('Rollno').count()
grad_count.reset_index(inplace=True)
stu_gpa=pd.DataFrame({"Name":grades.merge(students)['Name'].unique()})
stu_gpa['GPA']=grad_sum['Gradepoint']/grad_count['Gradepoint']
stu_gpa
```

## Question 6
Most generous teacher(s) -- highest mean GPA in course: (teacher name, course name)

```python
gd=grades.merge(courses).merge(instructors , left_on='Instructor', right_on='Instcode', suffixes=('_x','')).drop(columns='Instructor_x')
#gd=gd.groupby('Instructor')

grad_mean=gd.merge(gradepoint, left_on='Grade', right_on='Lettergrade').groupby('Instructor').mean()
grad_mean.reset_index(inplace=True)

inst_grades=pd.DataFrame({"Teacher Name":grad_mean['Instructor'], "GPA": grad_mean['Gradepoint']})

mean_high=inst_grades[inst_grades['GPA']==inst_grades['GPA'].max()]
mean_high=mean_high.merge(instructors, left_on='Teacher Name', right_on='Instructor').merge(courses, left_on='Instcode', right_on='Instructor', suffixes=('_x','')).drop(columns='Instructor_x')
mean_high.rename(columns={'Name': 'Course Name'}, inplace=True)
mean_high[['Teacher Name', 'Course Name']]
```

## Question 7
Most generous teacher(s) -- highest median grade in course: (teacher name, course name)

```python
gd=grades.merge(courses).merge(instructors , left_on='Instructor', right_on='Instcode', suffixes=('_x','')).drop(columns='Instructor_x')
#gd=gd.groupby('Instructor')

grad_median=gd.merge(gradepoint, left_on='Grade', right_on='Lettergrade').groupby('Instructor').median()
grad_median.reset_index(inplace=True)

inst_grades=pd.DataFrame({"Teacher Name":grad_median['Instructor'], "GPA": grad_median['Gradepoint']})

median_high=inst_grades[inst_grades['GPA']==inst_grades['GPA'].max()]
median_high=median_high.merge(instructors, left_on='Teacher Name', right_on='Instructor').merge(courses, left_on='Instcode', right_on='Instructor', suffixes=('_x','')).drop(columns='Instructor_x')
median_high.rename(columns={'Name': 'Course Name'}, inplace=True)
median_high[['Teacher Name', 'Course Name']]
```

## Question 8

Student(s) with overall highest GPA: (student name, GPA)

```python
stu=grades.merge(students).merge(gradepoint, left_on='Grade', right_on='Lettergrade')
grad_mean=stu.groupby('Name').mean()
grad_mean.reset_index(inplace=True)

stu_gpa=pd.DataFrame({"Name":grad_mean['Name'], "GPA":grad_mean['Gradepoint']})
stu_gpa[stu_gpa['GPA']==stu_gpa['GPA'].max()]
```

## Question 9
Student(s) with highest GPA in each house: (student name, house name, GPA)


```python
stu_hou=grades['Rollno'].str.slice(0,3,1)
grad_hou=grades.copy()
grad_hou['Housecode']=stu_hou
grad_hou=grad_hou.merge(houses).merge(students).merge(gradepoint, left_on='Grade', right_on='Lettergrade')
grad_mean=grad_hou.groupby(by=['Housename','Name']).mean()
house_max=grad_mean.groupby(level=0).max().reset_index(inplace=False)
grad_mean.reset_index(inplace=True)
stu_house_max=grad_mean.merge(house_max).rename(columns={"Gradepoint":"GPA", "Name":"Student Name"})
stu_house_max[['Student Name', 'Housename','GPA']]
```

## Question 10
Mean GPA of students in each house: (house name, GPA)

```python
grad_mean=grades.merge(gradepoint, left_on='Grade', right_on='Lettergrade').groupby('Rollno').mean()
grad_mean.reset_index(inplace=True)
grad_mean['Housecode']=grad_mean['Rollno'].str.slice(0,3,1)


grad_mean=grad_mean.merge(students).merge(houses)

hou_mean=grad_mean.groupby(by='Housename').mean()

hou_mean.reset_index(inplace=True)
hou_mean.rename(columns={"Gradepoint":"GPA"},inplace=True)
hou_mean[['Housename','GPA']]
```

## Question 11
Students in Slytherin with GPA higher than mean GPA in Slytherin: (student name)


```python
grad_mean=grades.merge(gradepoint, left_on='Grade', right_on='Lettergrade').groupby('Rollno').mean()
grad_mean.reset_index(inplace=True)
grad_mean['Housecode']=grad_mean['Rollno'].str.slice(0,3,1)
grad_mean=grad_mean.merge(students).merge(houses)
hou_mean=grad_mean.groupby(by='Housename').mean()
hou_mean.reset_index(inplace=True)
hou_mean.rename(columns={"Gradepoint":"GPA"},inplace=True)


sly_mean=hou_mean[hou_mean['Housename']=='Slytherin']['GPA']


grad_mean=grad_mean[grad_mean['Gradepoint']>float(sly_mean)]
grad_mean.rename(columns={"Name" :"Student Name"}, inplace=True)

grad_mean[grad_mean['Housename']=="Slytherin" ][['Student Name']]
```

## Question 12

Students in Hufflepuff with GPA higher than mean GPA in Gryffindor: (student name)

```python
grad_mean=grades.merge(gradepoint, left_on='Grade', right_on='Lettergrade').groupby('Rollno').mean()
grad_mean.reset_index(inplace=True)
grad_mean['Housecode']=grad_mean['Rollno'].str.slice(0,3,1)
grad_mean=grad_mean.merge(students).merge(houses)
hou_mean=grad_mean.groupby(by='Housename').mean()
hou_mean.reset_index(inplace=True)
hou_mean.rename(columns={"Gradepoint":"GPA"},inplace=True)


gryf_mean=hou_mean[hou_mean['Housename']=='Gryffindor']['GPA']


grad_mean=grad_mean[grad_mean['Gradepoint']>float(gryf_mean)]
grad_mean.rename(columns={"Name" :"Student Name"}, inplace=True)

grad_mean[grad_mean['Housename']=="Hufflepuff" ][['Student Name']]
```
